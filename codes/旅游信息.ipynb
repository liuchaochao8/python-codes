```python
import pandas as pd
data = pd.read_csv('D:\liuchaochao\桌面\大三上\数据分析\数据分析实验-第八次\疫情背景下的周边游需求图谱分析-测试数据\测试数据.csv',
                   encoding = 'utf-8')
data1 = pd.read_csv(r'D:\liuchaochao\桌面\大三上\数据分析\数据分析实验-第八次\疫情背景下的周边游需求图谱分析-测试数据\result1.csv',
                    encoding = 'utf-8')

judge = []
key = ['旅游' ,'活动' ,'节庆' ,'特产' ,'交通' ,'酒店' ,'景区' ,
       '景点' ,'文创' ,'文化' ,'乡村旅游' ,'民宿' ,'假日' ,
       '假期' ,'游客' ,'采摘' ,'赏花' ,'春游' ,'踏青' , 
       '康养' ,'公园' ,'滨海游' ,'度假' ,'农家乐' ,'剧本杀' ,
       '旅行' ,'徒步' ,'工业旅游' ,'线路' ,'自驾游' , 
       '团队游' ,'攻略' ,'游记' ,'包车' ,'玻璃栈道' ,'游艇' ,'高尔夫' ,'温泉']

oridata = list(data['正文'].astype(str))

for od in oridata:
    count = 0
    for k in key:
        if(k in od):
            #print(od)
            judge.append('符合')
            break
        else:
            count +=1
            
    if(count==38):
        judge.append('不符合')
        
data1['分类标签'] = judge
data1
```
