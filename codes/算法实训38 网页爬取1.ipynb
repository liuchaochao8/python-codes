# 编写一个Python程序来下载最新的校内通知网页，并保存为本地文件。

```python
# 1.使用requests库发送GET请求获取网页内容。
# 2.程序应该具有异常处理机制，能够处理网络请求失败的情况。
import requests
from requests import ReadTimeout
from bs4 import BeautifulSoup

def get_page(url,data):
    try:
        response = requests.get(url, data,timeout=1,json=True)
        if response.status_code == 200:
            print("Get Page Successfully")
            return response.text
        else:
            print('Get Page Failed', response.status_code)
            return None
    except (ConnectionError, ReadTimeout):
        print('Crawling Failed', url)
        return None
    
# 3.下载的网页内容应该保存为以网页标题命名的TXT文件，确保文件名合法。
def get_p_name(url,text):
    t = BeautifulSoup(text, 'lxml')
    print(t.title.text)
    
# 5.程序应该提供用户输入保存文件夹路径的功能。    
def get_path():
    filename = input("please input filename: ")
    print('InPut FileName: %s' % filename)
    return filename

# 4.下载的文件应该保存在指定的文件夹中。
# 6.下载完成后，应该输出提示信息表示下载成功，并显示保存的文件路径。
def save(fn,r,p_name):
    try:
        with open (fn+p_name+'.txt','w',encoding = 'utf8') as f:
            f.write(r)
        print('Successfully Write In!\n')
        print('Write In File: ',fn+p_name+'.txt')
    except:
        print('Failed!\n')
```
