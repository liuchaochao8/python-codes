```python
import pandas as pd
import re
```

```python
path = "D:\liuchaochao\桌面\zpt.csv"
df = pd.read_csv(path,low_memory=False,encoding ='gbk',header=None)
df.columns=['zpdate','job_name','com_name','job_loc','wage','welfare','com_type','require']
df['zpdate']=pd.to_datetime(df['zpdate'],format="%Y/%m/%d %H:%M")
df.head()
```

```python
jp = []
for i in range(df.shape[0]):
    st = df.iloc[i,3]
    sta = st[0:2]
    jp.append(sta)
    
df['job_province']=jp
df
```

```python
jw = []
for i in range(df.shape[0]):
    l = str(df.iloc[i][5])
    w = l.split()
    jw.append(w)
    
df['job_wel_list'] = jw    
df
```

```python
jn = []
for i in range(df.shape[0]):
    l = str(df.iloc[i][7])
    w = l.split()
    del w[0]
#     print(w)
    jn.append(w)
    
    
df['job_need_list'] = jn
df
```

```python
#去掉wage上的空值
print(df.isnull().sum(),'\n')
df.dropna(subset=['wage'],inplace=True)
print(df.isnull().sum())
```
