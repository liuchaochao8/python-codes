```python
import pandas as pd
import csv

#加载excel数据
dt_excel_1 = pd.read_excel('新冠疫情分析数据.xlsx')

Diagnosed={}
heal={}
death={}
city=[]

with open("task1_1.csv","w", newline='') as csvfile: 
    writer = csv.writer(csvfile) 
    writer.writerow( ["城市","日期","累计确诊人数","累计治愈人数","累计死亡人数"])
    date=dt_excel_1.iloc[0,0]
    for index,row in dt_excel_1.iterrows():
        if(row['日期'] != date):
            for i in city:
                writer.writerow([i,date,Diagnosed[i],heal[i],death[i]])
            date = row['日期']
        if row['城市'] not in city:
            city.append(row['城市'])
            Diagnosed[row['城市']] = row['新增确诊']
            heal[row['城市']] = row['新增治愈']
            death[row['城市']] = row['新增死亡']
        else:
            Diagnosed[row['城市']] += row['新增确诊']
            heal[row['城市']] += row['新增治愈']
            death[row['城市']] += row['新增死亡']
```

```python
import matplotlib.pyplot as plt
import pandas as pd
dt_excel_3 = pd.read_excel('D:\\liuchaochao\\桌面\\大三上\\excel实验\\新冠疫情分析数据-附件1.xlsx',sheet_name=3)

x = []
y = []

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

date = dt_excel_3.iloc[0,1]
for index,row in dt_excel_3.iterrows():
    if(row['通报日期'] != date):
        plt.scatter(x, y, color='green', s=5, label='疫情场所', alpha = 0.8)
        plt.title(f"{row['通报日期']-1}", fontdict={'size': 20})
        plt.xlabel("km", fontdict={'size': 16})
        plt.ylabel("km", fontdict={'size': 16})
        plt.legend(loc="upper right", prop={'size': 16})
        for i in range(len(x)):
            cir1 = plt.Circle((x[i], y[i]), 1, color="green", fill = False)
            plt.gcf().gca().add_artist(cir1)
        plt.show()
        date = row['通报日期']
    x.append(row['横坐标（公里）'])
    y.append(row['纵坐标（公里）'])
```
