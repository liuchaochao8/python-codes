# 预测一

## 思路:将前3年这一天的总功求平均

### 导入原始数据与待预测数据

```python
import datetime
import pandas as pd
dateparse = lambda dates: datetime.datetime.strptime(dates, '%Y/%m/%d %H:%M')
dateparse1 = lambda dates: datetime.datetime.strptime(dates, '%Y/%m/%d %H:%M\t')
data = pd.read_csv('D:\liuchaochao\桌面\大三上\数据分析\电力系统\电力系统负荷预测分析-全部数据\附件1-区域15分钟负荷数据.csv',
                   parse_dates=['数据时间'], 
                   #index_col='数据时间',#设置时间为索引
                   date_parser=dateparse)

unknow1 = pd.read_csv('D:\liuchaochao\桌面\大三上\数据分析\电力系统\电力系统负荷预测分析-测试数据\附件1.1-区域15分钟负荷预测数据.csv',
                   parse_dates=['数据时间'], 
                   #index_col='数据时间',#设置时间为索引
                   date_parser=dateparse1,
                    encoding='utf-8')

print (data.head())
print(unknow1.head())
```

### 拆分时间

```python
data['Year'] = data['数据时间'].dt.year
data['Month'] = data['数据时间'].dt.month
data['Day'] = data['数据时间'].dt.day
data['Hour'] = data['数据时间'].dt.hour
data['Minute'] = data['数据时间'].dt.minute

unknow1['Year'] = unknow1['数据时间'].dt.year
unknow1['Month'] = unknow1['数据时间'].dt.month
unknow1['Day'] = unknow1['数据时间'].dt.day
unknow1['Hour'] = unknow1['数据时间'].dt.hour
unknow1['Minute'] = unknow1['数据时间'].dt.minute

data.head()
#unknow1
```

### 待预测值为前三年当天数据平均

```python
prelist = []
for i in range(0,unknow1.shape[0]):
    avg = data.loc[(data['Month']==unknow1.iloc[i][3])&
              (data['Day']==unknow1.iloc[i][4])&
              (data['Hour']==unknow1.iloc[i][5])&
              (data['Minute']==unknow1.iloc[i][6]),['总有功功率（kw）']].mean()
    #print(avg[0])
    prelist.append(avg[0])
    

unknow1['总有功功率（kw）'] = prelist
unknow1
```
