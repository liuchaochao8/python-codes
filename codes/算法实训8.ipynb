```python
import numpy as np
import matplotlib.pyplot as plt
#令x为-5到5之间，元素数为100的等差数列
x = np.linspace(-5,5,100)
#输入直线方程
y = 0.5*x + 3
plt.plot(x,y,c='orange')
#图题设为"Straight Line"
plt.title('Straight Line')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [[1, 4]] 
y = [[3, 5]]

for i in range(len(x)):
    
    plt.plot(x[i], y[i], color='r')
    plt.scatter(x[i], y[i], color='b')
    plt.show()
```

```python
x = [1,4]
y = [3,5]
a = (x[0]-x[1])/(y[0]-y[1])
b = y[1]-x[1]*a

x = np.linspace(-5,5,100)
y = a*x + b
plt.plot(x,y,c='orange')
```
