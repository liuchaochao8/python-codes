```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

# 加载数据
iris = load_iris()
X = iris.data[:,:2]
y = iris.target

# 拆分数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 定义分类器
mlp = MLPClassifier(hidden_layer_sizes=(5,2), activation='relu', alpha=0.1)

# 训练模型
mlp.fit(X_train, y_train)

# 预测测试集
y_pred = mlp.predict(X_test)

# 输出准确率
print("Accuracy:", mlp.score(X_test, y_test))

# 绘制决策边界
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# 绘制训练集和测试集数据点
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title("hidden_layer_sizes=(5,2), activation='relu', alpha=0.1")

plt.show()
```

```python
# 定义分类器
mlp_1 = MLPClassifier(hidden_layer_sizes=(10,10,10), activation='tanh', alpha=0.1)

# 训练模型
mlp_1.fit(X_train, y_train)

# 预测测试集
y_pred = mlp_1.predict(X_test)

# 输出准确率
print("Accuracy:", mlp_1.score(X_test, y_test))

# 绘制决策边界
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = mlp_1.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# 绘制训练集和测试集数据点
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title("hidden_layer_sizes=(10,10,10), activation='tanh', alpha=0.1")

plt.show()
```

```python
models = (MLPClassifier(hidden_layer_sizes=(10,10), activation='tanh', alpha=0.1),
         MLPClassifier(hidden_layer_sizes=(5,10), activation='tanh', alpha=0.1),
         MLPClassifier(hidden_layer_sizes=(10,5), activation='tanh', alpha=0.1),
         MLPClassifier(hidden_layer_sizes=(3,17), activation='tanh', alpha=0.1),
         MLPClassifier(hidden_layer_sizes=(17,17), activation='tanh', alpha=0.1),
         MLPClassifier(hidden_layer_sizes=(17,3), activation='tanh', alpha=0.1))

models = (clf.fit(X, y) for clf in models)
acc = []

titles = ('hidden_layer_sizes=(10,10)',
          'hidden_layer_sizes=(5,10)',
          'hidden_layer_sizes=(10,5)',
          'hidden_layer_sizes=(3,17)',
         'hidden_layer_sizes=(17,17)',
         'hidden_layer_sizes=(17,3)')

fig, sub = plt.subplots(3, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

for clf, title, ax in zip(models, titles, sub.flatten()):
    ax.contourf(xx, yy,clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape),
                  cmap=plt.cm.plasma, alpha=0.5)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.plasma, s=10, edgecolors='k')
    

    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Feature 0')
    ax.set_ylabel('Feature 1')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
    acc.append(clf.score(X_test, y_test))

plt.show()
print(acc)
```
