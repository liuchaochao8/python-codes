```python
求出列表lists中每个元素出现的次数
 
方法一：使用函数Counter，可以迅速获取list中每个元素出现的次数
 
from collections import Counter
arr=[1,2,5,1,1,5,6,3,3,2,2,4,8]
# arr=Counter(lists)
# print(arr)
def counter(arr):
    return Counter(arr)
print(counter(arr))  
两种写法的结果都是Counter({1: 3, 2: 3, 5: 2, 3: 2, 6: 1, 4: 1, 8: 1})
```
