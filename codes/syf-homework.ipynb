```python
import pandas as pd
df = pd.read_csv('D:\liuchaochao\桌面\huangdi.csv')
df
```

```python
for i in range(df.shape[0]):#去掉未知数据
    if df.iloc[i,4]=='未知' or df.iloc[i,4]==None:
        df.iloc[i,4] = -1
```

```python
df = df[df['life'].notna()]#去掉空值
df.isnull().any()
```

```python
df = df.astype({'life':int})  #类型转化
df.info()
```

```python
bj = []
for i in range(df.shape[0]):#df.shape 二维数组(行数,列数)
    if df.iloc[i,4]==-1:
        bj.append('0')
    elif df.iloc[i,4]<20:
        bj.append('寿命低')
    elif df.iloc[i,4]<40:
        bj.append('寿命偏短')
    elif df.iloc[i,4]<60:
        bj.append('寿命偏长')
    else:
        bj.append('寿命长')
bj
```

```python
df.insert(loc=4,column='bijiao',value = bj)#插入
df.head()
```

```python

def corr_plot(data,need_corr=True,name='',title=''):
    plt.subplots(figsize = (12,11))
    sns.heatmap(data.corr(method='spearman') if need_corr else data, 
                annot=True,
                cmap = 'RdBu',#色盘，可以理解为配色主题(在这里用Palette参数会报错)
                linewidths = 0.1,#热力矩阵之间的线宽
                linecolor='white',
                vmax = 0.9,#vmax,vmin, 图例(热力bar)中最大值和最小值的显示值，没有该参数时默认不显示
                square=True
    )
    plt.title(title if name==None else title+name,fontsize = 20)
    plt.tight_layout()
    plt.savefig(title+name+'.jpg',dpi=300)
    plt.show()


# data是dataframe类型
corr_plot(data,title = 'Feature Correlation')

# %%
data.index = pd.to_datetime(data.index)

# %%
plt.figure(figsize=(7,6))
plt.scatter(data['融资净买入(亿元)'],data['沪深300'])
plt.title('散点图')
plt.ylabel("融资净买入(亿元)")
plt.xlabel('沪深300涨跌幅')
plt.show()

# %%
plt.figure(figsize=(10,6))
plt.plot(data['融资净买入(亿元)'])
plt.title('融资净买入(亿元)')
plt.show()

# %%
data

# %%
plt.figure(figsize=(8,6))
plt.hist(data['沪深300'],bins=30)
plt.title("沪深300涨跌幅")
plt.vlines(0,0,90,linestyle='--', colors='red')
plt.show()

# %% [markdown]
# ## PCA

# %%
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
def PCA_n(X_scaled):#主成分分析
    estimator_pca = PCA(n_components=None)
    estimator_pca.fit(X_scaled)
    evr = estimator_pca.explained_variance_ratio_ 
    plt.figure(figsize=(8, 5))
    plt.plot(np.arange(1, len(evr) + 1), np.cumsum(evr*100), "-o")
    plt.title("Cumulative Explained Variance Ratio", fontsize=15)
    plt.xlabel("number of components",fontsize=15)
    plt.ylabel("explained variance ratio(%)",fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()

def pcaX(X_scaled,n):#数据降维
    pca = PCA(n_components=n)
    pca_X = pca.fit_transform(X_scaled)
    print('降维：',X_scaled.shape,'-->',pca_X.shape)
    return pd.DataFrame(pca_X)

PCA_n(data)

# %%
data_new = pcaX(data,0.999)

data_new.to_excel("D:/liuchaochao/桌面/降维后的数据.xlsx")

# %%


# %%



```
