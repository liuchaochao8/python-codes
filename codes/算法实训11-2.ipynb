```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#载入糖尿病情数据集，将数据集拆分成训练集和测试集，使用线性回归模型进行拟合。
from sklearn.datasets import load_diabetes
X, y = load_diabetes().data, load_diabetes().target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 8)
lr = LinearRegression().fit(X_train, y_train)
X.shape, y.shape
```

```python
#导入岭回归
from sklearn.linear_model import Ridge
#使用岭回归对数据进行拟合
ridge = Ridge().fit(X_train, y_train)
print('\n\n\n代码运行结果：')
print('==========\n')
print("岭回归的训练数据集得分：{:.2f}".format(ridge.score(X_train, y_train)))
print("岭回归的测试数据集得分：{:.2f}".format(ridge.score(X_test, y_test)))
print('\n==========')
print('\n\n\n')
```

```python
#绘制线性回归的系数作为对比
plt.plot(lr.coef_, 'o', label = 'linear regression')

plt.hlines(0,0, len(lr.coef_))
plt.legend()
```

```python
alphas = np.logspace(-5,5,100)  # -10 到 2 取100份
```

```python
coefs = []
for alpha in alphas:
    # 获取模型 设置参数
    rr = Ridge(alpha=alpha,fit_intercept=False)
    rr.fit(X,y)
    coefs.append(rr.coef_)
```

```python
plt.figure(figsize=(8,6))  #设置画布大小
plt.plot(alphas,coefs)  #绘图
# 设置坐标轴 不是以均匀的方式展示 设置x轴线 而是 以10的倍数来显示
plt.xscale('log')
```
