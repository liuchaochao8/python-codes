### 统计出现次数与总词数,并排序

```python
import jieba
jieba.setLogLevel(jieba.logging.INFO)


txt_20 = open("D:/liuchaochao/桌面/大三上/数据挖掘作业/txt/二十大.txt", "r", encoding='utf-8').read()
txt_19 = open("D:/liuchaochao/桌面/大三上/数据挖掘作业/txt/十九大.txt", "r", encoding='utf-8').read()
txt_18 = open("D:/liuchaochao/桌面/大三上/数据挖掘作业/txt/十八大.txt", "r", encoding='utf-8').read()
txt_add = open("D:/liuchaochao/桌面/大三上/数据挖掘作业/合并.txt", "r").read()
    
word_20 = jieba.lcut(txt_20)
word_19 = jieba.lcut(txt_19)
word_18 = jieba.lcut(txt_18)
word_add = jieba.lcut(txt_add)

count_20 = {}
count_19 = {}
count_18 = {}
count_add = {}

sum20 = 0
sum19 = 0
sum18 = 0
sumadd = 0

#print(word_add)
#print("|".join(jieba.lcut(txt_add)))
```

```python
for word in word_20:
    if len(word) == 1:    # 单个词语不计算在内
        continue
    else:
        sum20 += 1
        count_20[word] = count_20.get(word, 0) + 1    # 遍历所有词语，每出现一次其对应的值加 1
item_20 = list(count_20.items())
item_20.sort(key=lambda x: x[1], reverse=True)    # 根据词语出现的次数进行从大到小排序    
    
    
    
for word in word_19:
    if len(word) == 1:    # 单个词语不计算在内
        continue
    else:
        sum19 += 1
        count_19[word] = count_19.get(word, 0) + 1    # 遍历所有词语，每出现一次其对应的值加 1
item_19 = list(count_19.items())
item_19.sort(key=lambda x: x[1], reverse=True)    # 根据词语出现的次数进行从大到小排序  



for word in word_18:
    if len(word) == 1:    # 单个词语不计算在内
        continue
    else:
        sum18 += 1
        count_18[word] = count_18.get(word, 0) + 1    # 遍历所有词语，每出现一次其对应的值加 1
item_18 = list(count_18.items())
item_18.sort(key=lambda x: x[1], reverse=True)    # 根据词语出现的次数进行从大到小排序


for word in word_add:
    if len(word) == 1:    # 单个词语不计算在内
        continue
    else:
        sumadd += 1
        count_add[word] = count_add.get(word, 0) + 1    # 遍历所有词语，每出现一次其对应的值加 1
item_add = list(count_add.items())
item_add.sort(key=lambda x: x[1], reverse=True)    # 根据词语出现的次数进行从大到小排序
```

```python
print(sum20,sum19,sum18,sumadd)
```

### 将文章拆分为句子

```python
import jieba
import re

txt_20 = txt_20.replace(u'\u3000',u'').replace('\n', '')#去掉文章中\n,\u3000
# 输入一个段落，分成句子，可使用split函数来实现
sentences = re.split(r'(。|！|\!|\.|？|\?)',txt_20)         # 保留分割符
sents_20 = []
for i in range(int(len(sentences)/2)):#合并句子与标点
    sent = sentences[2*i] + sentences[2*i+1]
    sents_20.append(sent)
    
    
txt_19 = txt_19.replace(u'\u3000',u'').replace('\n', '')#去掉文章中\n,\u3000
sentences = re.split(r'(。|！|\!|\.|？|\?)',txt_19)         # 保留分割符
sents_19 = []
for i in range(int(len(sentences)/2)):#合并句子与标点
    sent = sentences[2*i] + sentences[2*i+1]
    sents_19.append(sent)

txt_18 = txt_18.replace(u'\u3000',u'').replace('\n', '')#去掉文章中\n,\u3000
sentences = re.split(r'(。|！|\!|\.|？|\?)',txt_18)         # 保留分割符
sents_18 = []
for i in range(int(len(sentences)/2)):#合并句子与标点
    sent = sentences[2*i] + sentences[2*i+1]
    sents_18.append(sent)
    
    
txt_add = txt_add.replace(u'\u3000',u'').replace('\n', '')#去掉文章中\n,\u3000
sentences = re.split(r'(。|！|\!|\.|？|\?)',txt_add)         # 保留分割符
sents_add = []
for i in range(int(len(sentences)/2)):#合并句子与标点
    sent = sentences[2*i] + sentences[2*i+1]
    sents_add.append(sent)

sents_add
```

### 将句子拆分为单词

```python
from zhon.hanzi import punctuation
sen_word_20 = []
for i in range(len(sents_20)):
    for m in punctuation:
        sents_20[i] = sents_20[i].replace(m,'')
    word = []
    word = jieba.lcut(sents_20[i])
    for w in word:
        if(len(w)==1):
            word.remove(w)
    sen_word_20.append(word)

sen_word_19 = []
for i in range(len(sents_19)):
    for m in punctuation:
        sents_19[i] = sents_19[i].replace(m,'')
    word = []
    word = jieba.lcut(sents_19[i])
    for w in word:
        if(len(w)==1):
            word.remove(w)
    sen_word_19.append(word)
    
sen_word_18 = []
for i in range(len(sents_18)):
    for m in punctuation:
        sents_18[i] = sents_18[i].replace(m,'')
    word = []
    word = jieba.lcut(sents_18[i])
    for w in word:
        if(len(w)==1):
            word.remove(w)
    sen_word_18.append(word)
    
sen_word_add = []
for i in range(len(sents_add)):
    for m in punctuation:
        sents_add[i] = sents_add[i].replace(m,'')
    word = []
    word = jieba.lcut(sents_add[i])
    for w in word:
        #print(w,len(w))
        if(len(w)==1):
            word.remove(w)
    sen_word_add.append(word)

sen_word_add
```

```python
from mlxtend.preprocessing import TransactionEncoder  
from mlxtend.frequent_patterns import apriori  
from mlxtend.frequent_patterns import association_rules  
import pandas as pd  

data_set = sen_word_20[0:len(sen_word_20)]
    
#转换为算法可接受模型（布尔值）  
te = TransactionEncoder()  
df_tf = te.fit_transform(data_set)  
df = pd.DataFrame(df_tf,columns=te.columns_)  
#设置支持度求频繁项集  
frequent_itemsets = apriori(df,min_support=0.1,use_colnames= True) 
frequent_itemsets


items = []
for i in range(len(frequent_itemsets)):
    it = []
    for s in frequent_itemsets.iloc[i][1]:
        it.append(s)
    items.append(it)

#print(items)
frequent_itemsets['items'] = items


#按照 support 排序,并重置索引
frequent_itemsets = frequent_itemsets.sort_values(by="support",ascending=False)
frequent_itemsets.reset_index(drop=True,inplace=True)
print(frequent_itemsets)

#求关联规则,设置最小置信度 
rules = association_rules(frequent_itemsets,metric = 'confidence',min_threshold = 0.5)  
#设置最小提升度  
rules = rules.drop(rules[rules.lift <1.0].index)  
#设置标题索引并打印结果  
rules.rename(columns = {'antecedents':'from','consequents':'to','support':'sup','confidence':'conf'},inplace = True)  
rules = rules[['from','to','sup','conf','lift'] ]
#按照 lift 排序,并重置索引
rules = rules.sort_values(by="lift",ascending=False)
rules.reset_index(drop=True,inplace=True)

rules.head(20)
```

```python
writer = pd.ExcelWriter(r'D:/liuchaochao/桌面/大三上/数据挖掘作业/二十大分析结果.xlsx')
frequent_itemsets.drop(columns = frequent_itemsets.columns[[1]],axis=1, inplace=True)
frequent_itemsets.to_excel(writer, '频繁项集',index = None)
rules.to_excel(writer, '关联规则',index = None)
writer.close()
```
