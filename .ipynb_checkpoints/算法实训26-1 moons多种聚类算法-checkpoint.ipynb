```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
blobs = make_moons(n_samples=200,random_state=1)
X_blobs = blobs[0]
plt.scatter(X_blobs[:,0],X_blobs[:,1],c='r',edgecolor='k')
plt.show()
```

# kmeans

```python
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=5)
kmeans.fit(X_blobs)

#下面是用来画图的代码
x_min, x_max = X_blobs[:, 0].min()-0.5 , X_blobs[:, 0].max()+0.5 
y_min, y_max = X_blobs[:, 1].min()-0.5 , X_blobs[:, 1].max()+0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, .02), 
                     np.arange(y_min, y_max, .02))


Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.summer,
           aspect='auto', origin='lower')

plt.plot(X_blobs[:, 0], X_blobs[:, 1], 'r.', markersize=5)
#用蓝色叉号代表聚类的中心
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=150, linewidths=3,
            color='b', zorder=10)
plt.title("KMeans")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

# Spectral Clustering

谱聚类是从图论中演化出来的算法，它的主要思想是把所有的数据看做空间中的点，这些点之间可以用边连接起来。距离较远的两个点之间的边权重值较低，而距离较近的两个点之间的边权重值较高，通过对所有数据点组成的图进行切图，让切图后不同的子图间边权重和尽可能的低，而子图内的边权重和尽可能的高，从而达到聚类的目的。

```python
from sklearn.cluster import SpectralClustering
import numpy as np
data, target = make_moons(n_samples=200,random_state=1)
clustering = SpectralClustering(n_clusters=5,
        assign_labels='discretize',
        random_state=0).fit(data)
#下面是用来画图的代码
x_min, x_max = data[:, 0].min()-0.5 , data[:, 0].max()+0.5 
y_min, y_max = data[:, 1].min()-0.5 , data[:, 1].max()+0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, .02), 
                     np.arange(y_min, y_max, .02))
Z = clustering.fit_predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.summer,
           aspect='auto', origin='lower')
plt.plot(data[:, 0], data[:, 1], 'y.', markersize=5)
#用蓝色叉号代表聚类的中心
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=150, linewidths=3,
            color='b', zorder=10)
plt.title("SpectralClustering")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
cnames = ['black', 'blue', 'red', 'yellow', 'green', 'purple', 'orange']
for point, gp_id in zip(data, cluster_group.labels_):
    # 放到 plt 中展示
    plt.scatter(point[0], point[1], s=70, c=cnames[gp_id], alpha=1)

plt.show()
```
