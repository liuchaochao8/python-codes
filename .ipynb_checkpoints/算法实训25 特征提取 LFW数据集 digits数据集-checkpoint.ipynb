```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
faces = fetch_lfw_people(min_faces_per_person=20, resize=0.8)
image_shape = faces.images[0].shape
fig, axes = plt.subplots(2,4,figsize=(12,6),
                        subplot_kw={'xticks':(),'yticks':()})
for target,image,ax in zip(faces.target,faces.images,axes.ravel()):
    ax.imshow(image, cmap=plt.cm.gray)
    ax.set_title(faces.target_names[target])
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(faces.data/255, 
                                                    faces.target,
                                                   random_state=62)
mlp=MLPClassifier(hidden_layer_sizes=[100,100], random_state=62,
                  max_iter=400)
mlp.fit(X_train, y_train)
print(':{:.2f}'.format(mlp.score(X_test, y_test)))
```

```python
from sklearn.decomposition import PCA
pca = PCA(whiten=True, n_components=0.9, random_state=62).fit(X_train)
X_train_whiten = pca.transform(X_train)
X_test_whiten = pca.transform(X_test)
print('{}'.format(X_train_whiten.shape))
```

```python
mlp.fit(X_train_whiten, y_train)
print(':{:.2f}'.format(mlp.score(X_test_whiten, 
                                             y_test)))
```

```python
from sklearn.decomposition import NMF
nmf = NMF(n_components=105,random_state=62).fit(X_train)
X_train_nmf = nmf.transform(X_train)
X_test_nmf = nmf.transform(X_test)
print('NMF{}'.format(X_train_nmf.shape))
```

```python
mlp.fit(X_train_nmf, y_train)
print('nmf{:.2f}'.format(mlp.score(X_test_nmf,
                                           y_test)))
```

# digits

## PCA


```python
from sklearn.datasets import load_digits
# from sklearn.preprocessing import StandardScaler
# from sklearn.utils import check_random_state
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
#digits
dig = load_digits()
#
print(dig.keys())
print(dig.data.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dig.data/255, 
                                                    dig.target,
                                                   random_state=60)
mlp=MLPClassifier(hidden_layer_sizes=[100,100], random_state=62,max_iter=400)
mlp.fit(X_train, y_train)
print(':{:.5f}'.format(mlp.score(X_test, y_test)))
```

## PCA

```python
from sklearn.decomposition import PCA
#2
pca = PCA(n_components=20)
pca.fit(dig.data)
X_pca = pca.transform(dig.data)
print(X_pca.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, 
                                                    dig.target,
                                                   random_state=60)
mlp=MLPClassifier(hidden_layer_sizes=[100,100], random_state=62,max_iter=400)
mlp.fit(X_train, y_train)
print(':{:.5f}'.format(mlp.score(X_test, y_test)))
```

### PCA

```python
for k in range(1,65,2):
    pca = PCA(n_components=k)
    pca.fit(dig.data)
    X_pca = pca.transform(dig.data)
    #print(X_pca.shape)
    X_train, X_test, y_train, y_test = train_test_split(X_pca, 
                                                    dig.target,
                                                   random_state=60)
    mlp=MLPClassifier(hidden_layer_sizes=[100,100], random_state=62,max_iter=400)
    mlp.fit(X_train, y_train)
    print('n_components {:d} '.format(k)+':{:.5f}'.format(mlp.score(X_test, y_test)))
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components= 9)
lda.fit(dig.data,dig.target)
X_lda = lda.transform(dig.data)
print(X_lda.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_lda, 
                                                    dig.target,
                                                   random_state=60)
mlp=MLPClassifier(hidden_layer_sizes=[100,100], random_state=62,max_iter=400)
mlp.fit(X_train, y_train)
print(':{:.5f}'.format(mlp.score(X_test, y_test)))
```
