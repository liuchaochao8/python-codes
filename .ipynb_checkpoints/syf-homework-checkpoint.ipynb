```python
import pandas as pd
df = pd.read_csv('D:\liuchaochao\桌面\huangdi.csv')
df
```

```python
for i in range(df.shape[0]):#去掉未知数据
    if df.iloc[i,4]=='未知' or df.iloc[i,4]==None:
        df.iloc[i,4] = -1
```

```python
df = df[df['life'].notna()]#去掉空值
df.isnull().any()
```

```python
df = df.astype({'life':int})  #类型转化
df.info()
```

```python
bj = []
for i in range(df.shape[0]):#df.shape 二维数组(行数,列数)
    if df.iloc[i,4]==-1:
        bj.append('0')
    elif df.iloc[i,4]<20:
        bj.append('寿命低')
    elif df.iloc[i,4]<40:
        bj.append('寿命偏短')
    elif df.iloc[i,4]<60:
        bj.append('寿命偏长')
    else:
        bj.append('寿命长')
bj
```

```python
df.insert(loc=4,column='bijiao',value = bj)#插入
df.head()
```

```python
import matplotlib.pyplot as plt#画图
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('年龄统计')
df['bijiao'].value_counts().plot.bar()
p = sns.countplot(df['bijiao'])
```
